# Using PCA for overfitting


```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA()
principalComponents = pca.fit_transform(X_scaled)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

```python
pca = PCA(n_components=60)
pc = pca.fit_transform(X_scaled)

pc = pd.DataFrame(pc,index=X_scaled.index)
pc
```

    New Training and Test Spilt

```python
from sklearn.model_selection import train_test_split


X_train2,X_test2,y_train2,y_test2 = train_test_split(pc,y,test_size = 0.3)
```

```python
xgb = XGBClassifier()
xgb.fit(X_train2,y_train2)
xgb_pred = xgb.predict(X_test2)
xgb_pred
```

```python
xgb.score(X_train2,y_train2)
xgb_score = accuracy_score(y_test2,xgb_pred)
xgb_score
```

```python
rf = RandomForestClassifier()
rf.fit(X_train2,y_train2)
rf_pred = rf.predict(X_test2)
rf_pred
```

```python
rf.score(X_train2,y_train2)
rf_score = accuracy_score(y_test2,rf_pred)
rf_score
```

```python
KNN = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
KNN.fit(X_train2, y_train2)
KNN_pred = KNN.predict(X_test2)
KNN_pred
```

```python
KNN_score = accuracy_score(y_test2,KNN_pred)
KNN_score
```

```python
from sklearn.model_selection import GridSearchCV
svm = SVC()
param_grid = {"kernel": ['rbf', 'sigmoid'],
                          "C": [0.1, 0.5, 1.0],
                          "random_state": [0, 100, 200, 300]}
svm_grid = GridSearchCV(svm, param_grid=param_grid, cv=5, verbose=3)
svm_grid.fit(X_train2, y_train2)
```
